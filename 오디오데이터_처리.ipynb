# 오디오데이터원리

```python
import numpy as np


def get_wave(freq, time, amplitude):
  '''
  amplitude=진폭, 반지름
  freq=주파수,1초당 진동횟수
  '''

  sample_rate = freq * 100
  time_sr = np.linspace(0, time, int(sample_rate * time))

  # 파동함수
  wave = amplitude * np.sin(2 * np.pi * freq * time_sr)
    
  return time_sr, wave


x, y = get_wave(440, 1, 1) 
```

```python
import matplotlib.pyplot as plt
plt.plot(x[0:100], y[0:100])
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy
sr = 1000 # sample rate
T = 2.0    # 초
t = numpy.linspace(0, T, int(T*sr), endpoint=False) # 시간 변수
f = 440
a = 0.5 # 진폭
x = a*numpy.sin(2*numpy.pi*f*t)# 440hz 사인파 함수

from IPython.display import Audio
Audio(x, rate=sr*T) 
```
