```python
import numpy as np


def get_wave(freq, duration, amplitude):
  '''
  duration=지속시간
  amplitude=진폭,반지름
  freq=주파수,1초에진동하는 횟수
  '''
  # 그래프를 그리기 위해 hz당 100개로 나눔 
  samplerate = freq * 100 

  # linspace함수로 0에서 duration까지 hz당 100개의 숫자를 뽑아냄
  t = np.linspace(0, duration, int(samplerate * duration))

  # 파동함수
  wave = amplitude * np.sin(2 * np.pi * freq * t)
    
  return t, wave


x, y = get_wave(440, 2, 1)
```

```python
import matplotlib.pyplot as plt
plt.plot(x[0:100], y[0:100])
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()
```
